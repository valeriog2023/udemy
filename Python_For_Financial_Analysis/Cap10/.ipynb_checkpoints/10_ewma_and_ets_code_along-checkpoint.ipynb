```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
%ls
```

```python
#
# Note that the column Month is missing a couple of datapoint so we can't use
# parse_dates=True when we load the csv
df1 = pd.read_csv("airline_passengers.csv",index_col="Month", parse_dates=True)
df = pd.read_csv("airline_passengers.csv",index_col="Month")
df.head()
```

```python
# as you see Month is not an index and it is not of type datetime
df1.index
```

```python
df.dropna(inplace=True)
df.index = pd.to_datetime(df.index)
df.index
```

```python
#
# Simple Moving Averages
# Note that if you just call rolling you get an object of type rolling...
# you need to call a function of the object (e,g. mean) to get a value from the time series
df['6-month-SMA'] = df["Thousands of Passengers"].rolling(window=6).mean()
df['12-month-SMA'] = df["Thousands of Passengers"].rolling(window=12).mean()
df.head()
```

```python
df.plot(figsize=(10,6))
```

```python
#
# Exponential Moving averages
# they take as paarementer either
#  - com: decay
#  - span (used here)
#  - half life
df['EWMA-6'] = df["Thousands of Passengers"].ewm(span=6).mean()
df['EWMA-12'] = df["Thousands of Passengers"].ewm(span=12).mean()
df[["Thousands of Passengers", "EWMA-6", "EWMA-12"]].plot(figsize=(10,6))
#
# Because the points are weighted and the ones close to the current date time are considered more,
# you can see that the outline of the EWMA-6 is smoother the more you go in the past and more more
# similar to the actual line the more you get close to the present
#
# The mathematical explanantion of the ewma and the different options are a bit beyond the scope of this however
# you just have that the value at time T is given by the sum of all the previous ones multiplied by a weight W_i
# ( and divided by the sum of weights up to that value i)
#
# adjust = True (default)
# the wieghts are given by (1-alpha)^(t-i) where alpha < 1 so for instance the first point (i.e. the oldest in time)
# is weighted (1-alpha)^t  which is smaller than the last point in the sum that is weighted only (1-alpha)
#
# if adjust = False the weight have a slightly different function but the concept is the same..
#
# HOWEVER WE CAN'T SPECIFY THE ALPHA DIRECTLY.. WE CAN ONLY USE ONE OF THE OPTIONS SPECIFIED ABOVE:
# - COM (CENTER OF MASS)  ---> ALPHA = 1/(1+C)
#   More Physical interpretation
# - SPAN --------------------> ALPHA = 2/(S+1)
#   Typically represents a N period moving average 
# - HALF LIFE ---------------> ALPHA = 1 - EXP( log0.5 / h)
#   Period of time for the exponential to get to half value
```

```python
#
#
# ETS Decomposition
# Error / Trend / Seasonality Decomposition
#
# IMPORT FROM STATSMODEL
from statsmodels.tsa.seasonal import  seasonal_decompose
#
# Original line
df["Thousands of Passengers"].plot()
```

```python
#
# run the seasonal decomposition
#
# Note that you can use different models for the decompose : additive or multiplicative
#      you can maybe guess from the original plot line which model is better..
#      (otherwise guess and check..)
#      above you can see that line is not too linear (just look at the peaks, if you connect them
#      you can see it's not really a line.. more like a parabola..
#      so the first guess would be multiplicative growth model
df.dropna(inplace=True)
df.index = pd.to_datetime(df.index)
result = seasonal_decompose(df["Thousands of Passengers"], model='multiplicative')
```
