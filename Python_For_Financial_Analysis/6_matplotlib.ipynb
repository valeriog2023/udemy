```python
import matplotlib.pyplot as plt
```

```python
# this makes you see the plot inline in the jupyter notebook
%matplotlib inline
```

```python
import numpy as np 
x = np.linspace(0,5,11)
y =  x ** 2
```

```python
#
# matplotlib has 2 wasy to crate plots:
# - functional
# - object based

#
# 1) functional
plt.plot(x,y)
plt.xlabel("X label")
plt.ylabel("Y label")
plt.title("Title1")
plt.show()    # not really required in jupyter
#
#
```

```python
#
# 2) functional use subplot
#   Parameters for subplot are number of rows, number of columns, which subplot you are working on
plt.subplot(1,2,1)  
plt.plot(x,y, 'r')
plt.title("sub1. title")
plt.xlabel("sub1. X")
plt.ylabel("sub1. Y")

plt.subplot(1,2,2)  
plt.plot(y,x,'b')
plt.title("sub2. title")
plt.xlabel("sub2. X")
plt.ylabel("sub2. Y")
```

```python
#
# OBJECT Oriented Approach
# we create a figure object and use methods
# the figure object is like a canvas
fig = plt.figure()
#
# add axes
# it takes 4 arguments, left, bottom, witdh, height that go from 0 to 1
axes = fig.add_axes([0.1,0.1, 0.8,0.8])
#
# now we can plot on the set of axes (and also set the other attributes: x label, y label, title)
axes.plot(x,y)
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("title")
fig.show()


```

```python
#
# OBJECT Oriented Approach
# we create a figure object and use methods
fig = plt.figure()
#
# add axes
# it takes 4 arguments, left, bottom, witdh, height that go from 0 to 1 (it's percentage)
# this time we add 2 set of axes
axes1 = fig.add_axes([0.1,0.1, 0.8,0.8])
axes2 = fig.add_axes([0.2,0.5, 0.4,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title("Title1")
axes2.set_title("Title2")
axes1.set_xlabel("Set 1. X")
axes2.set_xlabel("Set 2. X")
axes1.set_ylabel("Set 1. Y")
axes2.set_ylabel("Set 2. Y")
fig.show()
```

```python
#
# OBJECT Oriented Approach
# we create a figure object and use methods
# with subplots I can specify the number of rows and columns
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12,5))

#
# now we can use axes again which is an array of axes objects
# note that if you give more than one dimension, you get a matrix of axes..
for row_index,current_row_axes in enumerate(axes):
    for column_index, current_ax in enumerate(current_row_axes):
        current_ax.set_title(f"Plot {row_index}_{column_index}")
        current_ax.set_xlabel(f"Plot {row_index}_{column_index} - X")
        current_ax.set_ylabel(f"Plot {row_index}_{column_index} - Y")
        current_ax.plot(x,y)
        
plt.tight_layout()        
    
```

```python
#
# OBJECT Oriented Approach
# we create a figure object and use methods
fig = plt.figure()
#
# add axes
# it takes 4 arguments, left, bottom, witdh, height that go from 0 to 1 (it's percentage)
# this time we plot 2 lines in the same set of axes
axes = fig.add_axes([0.1,0.1, 0.8,0.8])
axes.plot(x,x**2, label = 'X squared')
axes.plot(x,x**3, label = 'X cubed')
axes.set_title("Title1")
axes.set_xlabel("Set 1. X")
axes.set_ylabel("Set 1. Y")
#
# we want to add a legend (it needs a label set in the matching plot command)
axes.legend()
fig.show()
plt.tight_layout()   
```

```python
#
#
#
# MISC Features
# - colors
#   this can take multiple parameters, e.g. color names, rgb hex code #a8cfcb
# - line width and style
# - alpha : controls how transparent the line is (0-1) 
# - line styles : example dashed, dotted, etc.. 
#    '-': 
#    '--' 
#    '-.' 
#    'steps'
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y, color='#a8cfcb', linewidth=4, alpha=0.8, linestyle='steps')
#
# markers: o, +, 1, ..
# with size, color, etc...
axes.plot(x,y, marker='o', markersize=20 ,
          markerfacecolor='yellow',
          markeredgewidth=3,
          markeredgecolor='green'
           )
#
# limits
# set limits to the x/y axis (remove to see how it was originally)
axes.set_xlim(3)
axes.set_ylim(10)
```
